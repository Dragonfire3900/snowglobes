```python

%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np

#plt.rcParams["font.family"] = "Times"
#font = {'family' : 'Times',
#        'weight' : 'bold',
#        'size'   : 22}

font = {'family' : 'Times',
        'size'   : 18}

plt.rc('font', **font)
#plt.style.use('seaborn')

detconfigname = "ar40kt"
fluxname = "livermore"
channelfile = "argon"

chanfilename = "../channels/channels_"+channelfile+".dat"
chans = np.genfromtxt(chanfilename,dtype={'names': ('cname', 'num', 'cpstate','flavor','ntargets'),'formats': ('U16', 'i', 'U12','U12','i')})

# Go through the channels in the channel file
ich=0
while ich<chans.size:
    #print(chans['cname'][ich],chans['cpstate'][ich])
    channame = chans['cname'][ich]
    #print(channame)
    smearedfilename = "../out/"+fluxname+"_"+str(channame)+"_"+detconfigname+"_events_smeared.dat"
    print(smearedfilename)
    smeared = np.loadtxt(smearedfilename, comments=['#', '--', 'Total'], skiprows=0)
    energyingev = smeared[:,0]
    energyinmev = energyingev*1000.

    smeared_events = smeared[:,1]
#    plt.plot(energyinmev,xscn_nue,label=r'$\bar{\nu}_e$')
    plt.plot(energyinmev,smeared_events,label=chans['cname'][ich])
    
    ich=ich+1


plt.xlabel("Observed Energy (MeV)")
plt.ylabel("Events per bin")

plt.legend(loc='upper right',prop={'size': 12})

```
