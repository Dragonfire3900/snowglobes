```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np

#plt.rcParams["font.family"] = "Times"
#font = {'family' : 'Times',
#        'weight' : 'bold',
#        'size'   : 22}

font = {'family' : 'Times',
        'size'   : 18}

plt.rc('font', **font)
#plt.style.use('seaborn')

fluxname = "garching_sum"

fluxfilename = "../fluxes/"+fluxname+".dat"

flux = np.loadtxt(fluxfilename, comments=['#', '--', 'Total'], skiprows=0)
energyingev = flux[:,0]
energyinmev = energyingev*1000.

eflux = flux[:,1]
muflux = flux[:,2]
tauflux = flux[:,3]
ebarflux = flux[:,4]
mubarflux = flux[:,5]
taubarflux = flux[:,6]
xflux = muflux+tauflux+mubarflux+taubarflux
#    plt.plot(energyinmev,xscn_nue,label=r'$\bar{\nu}_e$')
plt.xlim(0,50.)
plt.plot(energyinmev,eflux,label=r'${\nu}_e$',color='r')
plt.plot(energyinmev,ebarflux,label=r'$\bar{\nu}_e$',color='black')
plt.plot(energyinmev,xflux,label=r'${\nu}_x=\nu_\mu+\nu_\tau+\bar{\nu}_\mu+\bar{\nu}_\tau$',color='g')


plt.xlabel("Energy (MeV)")
# It's per bin, be careful
plt.ylabel("Neutrinos per 0.2 MeV per cm$^2$")

plt.legend(loc='upper right',prop={'size': 12})
figname = fluxname+".png"
plt.savefig(figname,dpi=300,bbox_inches='tight',facecolor="w")

```
